# **변수 설명**

## **타겟**
* tier: 플레이어의 티어(랭킹 수준).
* rank: 플레이어의 랭크 (티어 내의 세부 순위).

## **선택 변수**##
* champExperience: 챔피언이 얻은 총 경험치.
* champLevel: 챔피언 레벨.
* championName: 챔피언 이름.
* commandPings: 팀원에게 보낸 명령 핑의 개수.
* consumablesPurchased: 구매한 소모품(포션, 와드 등)의 개수.
* damageDealtToBuildings: 건물에 가한 총 피해량.
* damageDealtToObjectives: 오브젝트(드래곤, 바론 등)에 가한 총 피해량.
* damageDealtToTurrets: 타워에 가한 총 피해량.
* damageSelfMitigated: 플레이어가 완화한 총 피해량.
* dangerPings: 위험을 알리는 핑의 개수.
* deaths: 사망 횟수.
* detectorWardsPlaced: 설치한 탐지 와드의 개수.
* doubleKills: 더블 킬(연속 두 명 처치) 횟수.
* dragonKills: 플레이어가 처치한 드래곤 수.
* enemyMissingPings: 적 위치가 보이지 않음을 알리는 핑 개수.
* enemyVisionPings: 적의 시야 위치를 알리는 핑 개수.
* firstBloodAssist: 첫 킬을 지원했는지 여부.
* firstBloodKill: 첫 킬을 기록했는지 여부.
* firstTowerAssist: 첫 타워 파괴에 기여했는지 여부.
* firstTowerKill: 첫 타워를 파괴했는지 여부.
* getBackPings: 돌아가라는 핑의 개수.
* goldEarned: 획득한 총 골드.
* goldSpent: 사용한 총 골드.
* holdPings: 대기하라는 핑의 개수.
* individualPosition: 플레이어가 게임 내에서 실제로 플레이한 포지션
* inhibitorKills: 직접 억제기를 파괴한 횟수
* inhibitorTakedowns: 억제기를 파괴에 기여한 횟수
* inhibitorsLost: 잃은 억제기 수.
* item0 - item6: 장비 슬롯에 장착된 아이템 ID.
* itemsPurchased: 구매한 아이템의 총 개수.
* killingSprees: 킬 연속 횟수.
* kills: 총 처치 횟수.
* largestCriticalStrike: 최대 치명타 피해량.
* largestKillingSpree: 연속 킬 최대 횟수.
* largestMultiKill: 멀티 킬 중 최대 횟수.
* longestTimeSpentLiving: 생존한 가장 긴 시간.
* magicDamageDealt: 가한 마법 피해량.
* magicDamageDealtToChampions: 챔피언에게 가한 마법 피해량.
* magicDamageTaken: 받은 마법 피해량.
* needVisionPings: 시야 요청 핑의 개수.
* neutralMinionsKilled: 처치한 중립 미니언 수.
* objectivesStolen: 훔친 오브젝트 수(예: 드래곤, 바론 등).
* objectivesStolenAssists: 오브젝트 스틸 지원 횟수.
* onMyWayPings : 플레이어의 가는중 핑 횟수.
* physicalDamageDealt: 가한 물리 피해량.
* physicalDamageDealtToChampions: 챔피언에게 가한 물리 피해량.
* physicalDamageTaken: 받은 물리 피해량.
* pushPings: 전진하라는 핑의 개수.
* sightWardsBoughtInGame: 게임 내에서 구매한 시야 와드의 개수.
* spell1Casts : Q스킬 사용 횟수
* spell2Casts : W스킬 사용 횟수
* spell3Casts : E스킬 사용 횟수
* spell4Casts : R스킬 사용 횟수
* summoner1Casts : D 주문 사용 횟수
* summoner2Casts : F 주문 사용 횟수
* summoner1Id : D 주문 종류
* summoner2Id : F 주문 종류
* timeCCingOthers: 다른 챔피언을 군중 제어(CC)한 총 시간.
* timePlayed: 플레이한 총 시간.
* totalAllyJungleMinionsKilled: 아군 정글에서 처치한 미니언 수.
* totalDamageDealt: 총 가한 피해량.
* totalDamageDealtToChampions: 챔피언에게 가한 총 피해량.
* totalDamageShieldedOnTeammates: 팀원에게 제공한 방어막의 총 피해량.
* totalDamageTaken: 받은 총 피해량.
* totalEnemyJungleMinionsKilled: 적 정글에서 처치한 미니언 수.
* totalHeal: 회복한 총 체력량.
* totalHealsOnTeammates: 팀원에게 제공한 총 치유량입니다. (힐링 스킬,아이템)
* totalMinionsKilled: 게임 내에서 플레이어가 처치한 미니언의 총 수.
* totalTimeCCDealt: 적에게 가한 총 군중 제어(CC)시간 적을 스턴, 슬로우 등으로 무력화한 시간(초 단위)
* totalTimeSpentDead: 게임 중 플레이어가 사망하여 대기한 총 시간(초 단위)
* totalUnitsHealed: 플레이어가 회복시킨 총 유닛 수. 자신과 팀원을 포함한 치유 대상 수.
* trueDamageDealt: 가한 순수 피해량입니다. 방어 무시 피해. 방어력이나 마법 저항에 영향 X
* trueDamageDealtToChampions: 챔피언에게 가한 순수 피해량
* trueDamageTaken: 받은 순수 피해량. 방어력이나 마법 저항과 무관하게 받은 피해량.
* turretKills: 플레이어가 파괴한 타워의 수.
* turretTakedowns: 타워 파괴에 기여한 횟수. 최종 타격을 하지 않아도 기여한 경우.
* turretsLost: 아군 팀이 잃은 타워의 수.
* visionClearedPings: 시야 제거와 관련된 핑의 수.주로 상대 시야를 제거할 때 사용됩니다.
* visionScore: 플레이어의 시야 점수입니다. 와드 설치, 제거, 적 시야 방해 등을 통해 얻은 점수입니다.
* visionWardsBoughtInGame: 게임 내에서 구매한 제어 와드의 수입니다.
* wardsKilled: 파괴한 적 와드의 수입니다.
* wardsPlaced: 설치한 와드의 수입니다.
* win: 게임의 승리 여부입니다. 승리 시 True, 패배 시 False로 표시됩니다.
* allInPings: 전투 시작 또는 돌격을 알리는 핑의 수입니다.
* assistMePings: 지원 요청 핑의 수입니다.
* assists: 어시스트 횟수입니다. 적을 처치하는 데 기여한 횟수를 나타냅니다.
* basicPings: 플레이어가 사용한 기본 핑의 수입니다.
* bountyLevel: 플레이어의 현상금 레벨입니다. 킬을 연속으로 성공할 때 부여되는 현상금 단계입니다.
* controlWardsPlaced: 설치한 제어 와드(핑 와드)의 총 개수입니다.
* damagePerMinute: 분당 평균 피해량을 나타냅니다. 게임 내 총 피해량을 플레이 시간으로 나눈 값입니다.
* damageTakenOnTeamPercentage: 팀 전체가 받은 피해량 중에서 플레이어가 받은 피해량의 비율을 나타냅니다.
* earlyLaningPhaseGoldExpAdvantage: 초반 라인전에서 얻은 골드 및 경험치 우위를 나타냅니다.
* effectiveHealAndShielding: 플레이어가 제공한 총 회복 및 보호막량을 나타냅니다. 팀원을 힐하거나 보호막을 준 양을 포함합니다.
* enemyChampionImmobilizations: 적 챔피언을 무력화(CC)한 횟수를 나타냅니다.
* fullTeamTakedown: 팀 전체가 참여한 처치 횟수를 나타냅니다.
* goldPerMinute: 분당 골드 획득량입니다. 게임에서 플레이어가 얻은 평균 골드 수입을 의미합니다.
* immobilizeAndKillWithAlly: 아군과 협력하여 적을 무력화(이동 불가 상태로 만듦)하고 처치한 횟수를 의미합니다.
* kTurretsDestroyedBeforePlatesFall: 포탑 방어막이 유지되는 시간 동안 파괴한 타워의 수입니다.
* kda: 킬-데스-어시스트(Kill-Death-Assist) 비율입니다. (킬 + 어시스트) / 데스 값으로 계산되며, 플레이어의 생존율과 기여도를 나타냅니다.
* killAfterHiddenWithAlly: 아군과 함께 적에게 숨어 있다가 처치한 횟수입니다.
* killParticipation: 팀의 총 킬에서 플레이어가 관여한 비율입니다.
* killedChampTookFullTeamDamageSurvived: 적 챔피언에게 팀의 총 피해를 가한 후 생존한 횟수입니다.
* landSkillShotsEarlyGame: 초반에 적중한 스킬 샷의 횟수입니다.
* laneMinionsFirst10Minutes: 게임 시작 후 10분 이내에 처치한 라인 미니언 수입니다.
* laningPhaseGoldExpAdvantage: 라인전에서 얻은 골드와 경험치 우위를 나타냅니다.
* outnumberedKills: 적이 수적으로 우세한 상황에서 처치한 적의 수입니다.
* soloKills: 단독 킬 횟수입니다. 아군의 도움 없이 적을 처치한 경우를 나타냅니다.
* stealthWardsPlaced: 은신 와드(일반 와드)를 설치한 횟수입니다.
* takedownsInEnemyFountain: 적의 포탑 안에서 적을 처치하거나 어시스트한 횟수입니다.
* teamBaronKills: 팀이 처치한 바론의 수입니다.
* teamDamagePercentage: 팀 전체 데미지에서 플레이어가 차지하는 비율입니다. 플레이어의 기여도를 나타냅니다.
* teamElderDragonKills: 팀이 처치한 장로 드래곤의 수입니다.
* teamRiftHeraldKills: 팀이 처치한 리프트 헤럴드의 수입니다.
* turretPlatesTaken: 파괴한 타워 방어막의 수입니다. 타워 방어막은 게임 초반에 타워에 부여되는 보호막입니다.
* visionScoreAdvantageLaneOpponent: 라인 상대와 비교했을 때 시야 점수 우위를 나타냅니다.
* visionScorePerMinute: 분당 평균 시야 점수입니다.
* wardTakedowns: 파괴한 와드의 수입니다.
* wardTakedownsBefore20M: 게임 시작 후 20분 이내에 파괴한 와드의 수입니다.
* wardsGuarded: 플레이어가 방어한 와드의 수입니다. 팀의 시야를 유지하거나 보호한 횟수로, 와드를 적의 공격으로부터 지켜낸 경우를 나타냅니다.



## **부가적인 확인이 필요한 변수**
* summonerLevel: 소환사 계정 레벨.
* lane: 플레이어가 주로 플레이한 라인(탑, 미드 등).
* role: 게임에서 수행한 역할(예: 서포트, 딜러 등).
* nexusLost: 잃은 넥서스 수.(게임이 항복으로 끝났는가, 질질 끌다가 넥서스가 부숴졌는가.)
* abilityUses: 플레이어가 스킬을 사용한 총 횟수입니다. 모든 스킬과 주문 사용 횟수를 포함하지 않기 때문일 가능성이 높습니다. abilityUses는 특정 조건 하에서만 주요 스킬 사용 횟수를 집계
* deathsByEnemyChamps: 적 챔피언에 의해 사망한 횟수입니다.
* dodgeSkillShotsSmallWindow: 작은 시간 내에 회피한 스킬 샷의 수를 나타냅니다.
* enemyJungleMonsterKills: 적 정글에서 처치한 몬스터 수입니다.
* firstTurretKilled: 첫 번째 타워를 파괴했는지 여부입니다.
* firstTurretKilledTime: 첫 번째 타워가 파괴된 시간(초 단위)입니다.
* gameLength: 게임의 총 길이(초 단위)입니다. 한 판이 끝날 때까지의 총 시간을 나타냅니다.(모든 소환사 동일)
* timePlayed: 플레이한 총 시간.(접속 끊김,도주 등 실제 플레이 시간 기록)
* jungleCsBefore10Minutes: 게임 시작 후 10분 이내에 처치한 정글 미니언 수(CS)입니다.
* maxCsAdvantageOnLaneOpponent: 라인 상대와 비교한 최대 CS(Creep Score) 우위입니다. 라인전에서 적 챔피언보다 더 많은 미니언을 처치하여 얻은 최대 우위를 나타냅니다.
* legendaryItemUsed: 전설 아이템을 사용한 횟수 또는 전설 아이템 사용 여부를 나타냅니다.
* maxKillDeficit: 게임 내에서 한 번에 가장 많이 뒤처졌던 킬 격차입니다. 적이 얼마나 많은 킬을 앞섰는지의 최대치를 나타냅니다.
* maxLevelLeadLaneOpponent: 라인 상대보다 앞선 최대 레벨 차이를 나타냅니다.
* skillshotsDodged: 적의 스킬 샷을 회피한 횟수입니다.
* skillshotsHit: 적에게 적중시킨 스킬 샷의 횟수입니다.
* survivedThreeImmobilizesInFight: 한 전투에서 세 번의 무력화(이동 불가) 효과를 받고도 생존한 횟수입니다.

## **삭제**
* leaguePoints: 리그 포인트(LP).
* championId: 챔피언의 고유 ID.
* championTransform: 챔피언이 변신했는지 여부.
* gameEndedInSurrender: 게임이 항복으로 종료되었는지 여부.
* nexusKills: 넥서스 막타를 쳐서 파괴했는가.
* nexusTakedowns: 넥서스를 공격한 횟수.
* participantId: 플레이어의 고유 ID.
* pentaKills: 펜타킬 횟수(연속 5명 처치).
* quadraKills: 쿼드라킬 횟수(연속 4명 처치).
* profileIcon: 플레이어가 사용한 프로필 아이콘 ID.
* teamId: 100(블루),200(레드)
* teamPosition: 캐릭터 배정 당시의 포지션
* tripleKills: 트리플 킬(연속 3명 처치) 횟수입니다.
* unrealKills: 언리얼 킬 횟수.빠른 시간 안에 5명 이상을 처치한 경우 기록.
* baronKills: 플레이어가 막타로 처치한 바론의 수입니다.
* baronTakedowns: 바론을 처치에 기여한 횟수
* blastConeOppositeOpponentCount: 적을 맞은 반대편으로 날려버린 블라스트 콘(폭발식 식물) 사용 횟수입니다.
* 12AssistStreakCount: 12회 연속 어시스트를 기록한 횟수입니다. 플레이어가 연속으로 12번 이상 어시스트한 경우 카운트됩니다.
* InfernalScalePickup: 특정 아이템 또는 강화 요소와 관련된 변수일 가능성이 있지만, 구체적인 기능은 보통 패치나 모드에 따라 달라집니다.
* acesBefore15Minutes: 게임 시작 후 15분 이내에 발생한 에이스 횟수입니다.
* alliedJungleMonsterKills: 플레이어가 아군 정글에서 처치한 몬스터의 수입니다.
* bountyGold: 플레이어가 현재 현상금으로 가지고 있는 골드 양입니다. 킬이나 연속 처치로 인해 쌓인 현상금을 의미합니다.
* buffsStolen: 상대팀 정글에서 훔친 버프의 수입니다. 예를 들어 블루 버프나 레드 버프 등을 포함합니다.
* completeSupportQuestInTime: 서포트 퀘스트를 지정된 시간 내에 완료했는지 여부를 나타냅니다. 서포트 전용 퀘스트 아이템과 관련됩니다.
* dancedWithRiftHerald: 리프트 헤럴드 근처에서 춤을 춘 횟수를 나타냅니다. 주로 유머나 이벤트적인 요소로 포함될 수 있는 변수입니다.
* doubleAces: 양 팀이 동시에 에이스를 기록한 횟수입니다.
* dragonTakedowns: 플레이어가 처치한 드래곤의 수입니다. 최종 타격을 기록한 경우 포함됩니다.
